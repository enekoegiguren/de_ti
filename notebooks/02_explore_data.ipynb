#### Null values

```python
# Replacing '-' with NaN in the entire dataframe
df_gouv.replace('-', 0, regex=True, inplace=True)
```

```python
df_gouv.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Assuming df_gouv is your DataFrame
# df_gouv.isnull().sum() gives the count of missing values in each column
missing_values_count = df_gouv.isnull().sum()

# Plotting
plt.figure(figsize=(10, 6))
missing_values_count.plot(kind='bar', color='skyblue', alpha=0.7)
plt.title('Missing Values Count per Column')
plt.xlabel('Columns')
plt.ylabel('Missing Values Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
